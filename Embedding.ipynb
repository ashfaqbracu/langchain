```python
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(model="text-embedding-3-large")
embeddings
```

```python
text = " how are you doing today? I hope you are doing well. This is a test document. We will use it to test the HTML splitter. The HTML splitter is a useful tool for splitting HTML documents into smaller chunks. This is especially useful for large documents that need to be processed in smaller parts."

query_result = embeddings.embed_query(text)
query_result
```

```python
len(query_result)
```

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader('speech.txt')
docs = loader.load()

docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
texts = text_splitter.split_documents(docs) 
texts
```

```python
#### vector embedding and vector store

from langchain_community.vectorstores import FAISS

# Create FAISS vector store from documents
db = FAISS.from_documents(texts, embeddings)

# Display the vector store
db
```
